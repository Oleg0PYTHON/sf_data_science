```python

import pandas as pd

rm_df = pd.read_csv('data/ratings_movies.csv')
rm_df.head()
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
rm_df['year_release'] = rm_df['title'].apply(get_year_release)
rm_df.info()
```

```python
rm_df[rm_df['year_release'] == 2010].groupby(['genres'])['rating'].agg(['count', 'mean', 'max', 'min']).sort_values(by=['mean'], ascending=False)
```

```python
rm_df.groupby(['userId'])['genres'].agg(['count']).sort_values(by=['count'], ascending=False)
```

```python
rm_df.groupby(['userId'])['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[False, True])
```

```python
rm_df[rm_df['year_release'] == 2018].groupby(['genres'])['rating'].agg(['count', 'mean']).sort_values(by=['mean'], ascending=False)
```

```python
rm_df['date'] = pd.to_datetime(rm_df['date'])
rm_df['year_rating'] = rm_df['date'].dt.year
```

```python
pivot = rm_df.pivot_table(
    values='rating',
    index='genres',
    columns='year_rating',
    aggfunc=['mean'],
    fill_value=0
)
display(pivot.loc['Comedy'])
```

```python
display(pivot['mean'][2018].sort_values(ascending=False))
```

```python
merged = orders.merge(
    products,
    how='left',
    left_on = 'ID товара',
    right_on = 'Product_ID'
)
display(merged)
```

```python
merged['profit'] = merged['Количество']*merged['Price']
```

```python
merged[merged['Оплачен'] == 'Да'].groupby(['ID Покупателя'])['profit'].agg(['sum']).sort_values(by=['sum'], ascending=False)
```
